# Dimensionality Reduction and 3D Visualization of MNIST Dataset

## Introduction
This project aims to reduce the dimensionality of the MNIST dataset using PCA and t-SNE, and visually explore the clustering of digits in a 3D space. This can provide insights into the inherent structures of the dataset and the effectiveness of these reduction techniques.

## 1. Data Loading
Load the MNIST dataset which consists of 70,000 images of handwritten digits.

```python
from tensorflow.keras.datasets import mnist
import numpy as np

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
X = np.concatenate([train_images, test_images]).reshape(-1, 28*28)
y = np.concatenate([train_labels, test_labels])
# Normalize the data
X = X.astype('float32') / 255.0
```

## 3. Applying PCA and t-SNE for Dimensionality Reduction
Reduce the dimensions of the data to three dimensions using PCA and t-SNE.

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# PCA Reduction
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

# t-SNE Reduction
tsne = TSNE(n_components=3, verbose=1, perplexity=40, n_iter=300)
X_tsne = tsne.fit_transform(X)
```

## 4. 3D Visualization of the Reduced Data
Visualize the reduced data using 3D plots to observe the clustering of different digits.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# PCA Visualization
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(121, projection='3d')
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap='tab10')
plt.colorbar(scatter, ax=ax)
ax.set_title('PCA Reduction')

# t-SNE Visualization
ax = fig.add_subplot(122, projection='3d')
scatter = ax.scatter(X_tsne[:, 0], X_tsne[:, 1], X_tsne[:, 2], c=y, cmap='tab10')
plt.colorbar(scatter, ax=ax)
ax.set_title('t-SNE Reduction')
plt.show()
```
