```python
import pandas as pd

df = pd.read_csv('/content/datasets/Airbnb/airbnb.csv')
df.head()
```

```python
del df['img_links']
del df['host_id']

df.isnull().sum()
```

```python
df = df.dropna(subset=['name','address','safety_rules'])

df.isnull().sum()

df['amenities'].fillna(df['amenities'].mode()[0], inplace=True)
```

```python
df.dropna(subset=['name','address','safety_rules']).isnull().sum()
```

```python
df['reviews'] = df['reviews'].fillna('0')

df.dropna(subset=['name','address','safety_rules']).isnull().sum()
```

```python
df[df['rating'] != 'New'].isnull().sum()
```

```python
df['amenities']
```

```python
for i in df['features'][:10]:
  print(i)
```

```python
guests = []

for i in df['features']:

  flag = False

  for j in i.split(','):
    if 'guest' in j:
      flag = True
      guests.append(int(j.split(' ')[0]))

  if flag == False:
    guests.append(0)

df['guests'] = guests

df.head()
```
