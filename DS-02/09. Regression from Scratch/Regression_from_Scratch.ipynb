```python
years  = [1,2,3,4,5,6,7]
salary = [12,12,14,18,19,21,25]

w = []

for i,j in zip(years,salary):
  w.append(j/i)

w = sum(w)/len(w)
w
```

```python
import matplotlib.pyplot as plt

years  = [1,2,3,4,5,6,7]
salary = [12,12,14,18,19,21,25]

plt.plot(years, salary)  # Create a line plot
plt.xlabel("Years")  # Label the x-axis
plt.ylabel("Salary")  # Label the y-axis
plt.title("Salary vs. Years")  # Add a title
plt.show()  # Display the plot
```

```python
for x,y in zip(years,salary):
  print(x,x*w)
```

```python
years  = [1,2,3,4,5,6,7]
salary = [12,12,14,18,19,21,25]


x_values = []
y_values = []
for x, y in zip(years, salary):
  x_values.append(x)
  y_values.append(x * w)  # Calculate y values based on x*w


plt.plot(x_values, y_values, marker='o')  # Plot with markers for data points
plt.xlabel("Years")
plt.ylabel("Calculated Salary (x*w)")
plt.title("Calculated Salary vs. Years")
plt.show()
```

```python
import matplotlib.pyplot as plt

years  = [1,2,3,4,5,6,7]
salary = [12,12,14,18,19,21,25]

# Data for the first plot (original salary)
plt.plot(years, salary, label='Original Salary', marker='o')

# Data for the second plot (calculated salary)
x_values = []
y_values = []
for x, y in zip(years, salary):
  x_values.append(x)
  y_values.append(x * w)  # Calculate y values based on x*w
plt.plot(x_values, y_values, label='Calculated Salary (x*w)', marker='x')

# Customize the plot
plt.xlabel("Years")
plt.ylabel("Salary")
plt.title("Salary vs. Years")
plt.legend()  # Add a legend to distinguish the lines
plt.show()
```
