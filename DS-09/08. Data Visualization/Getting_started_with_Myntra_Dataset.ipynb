```python
import pandas as pd

df = pd.read_csv('/content/data.csv')

df.head()
```

### 1. EDA

##### 1.1) Checking the Null Values

```python
df.isnull().sum()
```

##### 1.2) Dropping the null values if any

```python
df.dropna()
```

##### 1.3) Dealing with Duplicates

```python
df.drop_duplicates(subset = 'product_link', inplace = True)
```

### 2. Data Cleaning

##### 2.1) Creating ID, Brand Tag & Description Column

```python
ids_              = []
product_category_ = []
brand_tag_        = []
description_      = []

for i in df['product_link']:

  product_category_.append(i.split('/')[0])
  brand_tag_.append(i.split('/')[1])
  description_.append(i.split('/')[2])
  ids_.append(i.split('/')[3])

df['product_id'] = ids_
df['product_category'] = product_category_
df['brand_tag'] = brand_tag_
df['description'] = description_

del ids_, product_category_, brand_tag_, description_

del df['sizes'], df['img_link'], df['discount_amount'], df['product_category']
del df['product_category'], df['brand_tag']

df = df[['product_id', 'product_name', 'brand_name', 'product_category', 'brand_tag',
         'description', 'product_tag', 'rating', 'rating_count', 'marked_price', 'discounted_price',
         'discount_percent', 'product_link']]

df.head()
```

##### 2.2) Checking unique products and removing if there are still any

```python
print(len(df['product_id']) == df['product_id'].nunique())

df = df.drop_duplicates(subset = ['product_id'])
```

##### 2.6) Deleting products with rating 0 or rating_count 0

```python
print('--before cleaning--')
print(round(df['rating'].mean(), 2))
print(round(df['rating_count'].mean(), 2))
print()
print()

print('--after cleaning-')
print(round(df[df['rating'] > 0]['rating'].mean(), 2))
print(round(df[df['rating_count'] > 0]['rating_count'].mean(), 2))

df = df[df['rating'] != 0]
```
