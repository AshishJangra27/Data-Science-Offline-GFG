### 2. Data Cleaning

```python
ids_              = []
product_category_ = []
brand_tag_        = []
description_      = []

for i in df['product_link']:

  product_category_.append(i.split('/')[0])
  brand_tag_.append(i.split('/')[1])
  description_.append(i.split('/')[2])
  ids_.append(i.split('/')[3])

df['product_id'] = ids_
df['product_category'] = product_category_
df['brand_tag'] = brand_tag_
df['description'] = description_

del ids_, product_category_, brand_tag_, description_

df.head()
```

```python
df[['product_category', 'brand_tag', 'description', 'product_id','buy']] = df['product_link'].str.split('/', expand=True)

del df['buy']

df.head()
```
